# TASK 2: Build Preprocessing Function

```python
import re

def preprocess_text(text):
    
    # handle empty or invalid input
    if not text or not isinstance(text, str):
        return [], ""

    # 1. convert to lowercase
    text = text.lower()

    # 2. remove URLs and emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # 3. remove numbers
    text = re.sub(r"\d+", "", text)

    # 4. handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # 5. remove special characters and emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # 6. remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # 7. tokenize
    tokens = text.split()

    # 8. remove short words (except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # final clean sentence
    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

# TASK 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, clean = preprocess_text(s)

    print("Original Text :", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 50)
```

# TASK 4: Token Analytics

```python
def analyze_tokens(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(word) for word in tokens) / total_tokens

    return total_tokens, unique_tokens, avg_length


print("TASK 4: TOKEN ANALYTICS\n")

for s in sentences:
    tokens, clean = preprocess_text(s)
    
    total, unique, avg = analyze_tokens(tokens)

    print("Sentence:", s)
    print("Tokens:", tokens)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg, 2))
    print("-" * 50)
```

# TASK 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

# collect all tokens
for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

# count frequency
counter = Counter(all_tokens)

print("TASK 5: FREQUENCY ANALYSIS\n")

# Top 10 frequent words
print("Top 10 Most Frequent Words:")
for word, count in counter.most_common(10):
    print(word, ":", count)

print("\nTop 5 Least Frequent Words:")
least_common = counter.most_common()[-5:]

for word, count in least_common:
    print(word, ":", count)
```

# TASK 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

# TASK 7: Error Handling

```python
test_cases = [
    "",                     # empty string
    "😂😂😍🔥",              # only emojis
    "1234567890",           # only numbers
    "     ",                # only spaces
]

print("TASK 7: ERROR HANDLING\n")

for t in test_cases:
    tokens, clean = preprocess_text(t)
    
    print("Input:", repr(t))
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print("-" * 50)
```
